```python
import csv
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob 
from nltk.corpus import wordnet
import time
import datetime
```

```python
def clean(tweet):
    tweet=(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()))
    return (tweet)
```

```python
all_stopwords = stopwords.words('english')

def remove_stopwords(tweet):
    tweet=word_tokenize(tweet)
    tweet= (" ".join([word for word in tweet if not word in all_stopwords]))
    return(tweet)

def get_tweet_sentiment(tweet):
        analysis = TextBlob(tweet)
        return analysis.sentiment.polarity

```

```python
def Datewise(citycsv):
    city = pd.read_csv(citycsv)
    tweet=city['text'].values.tolist()
    start_date=datetime.date(2020,6,19)
    end_date = datetime.date(2020,6,21)
    delta = datetime.timedelta(days=1)
    date_compare =[]
    while start_date<= end_date:
        date_compare.append(start_date)
        start_date+=delta
    
    day1=[]
    day2=[]
    day3=[]
    dates=city['date'].values.tolist()
    for i in range(len(dates)):
        if date_compare[0]==datetime.datetime.strptime(dates[i],"%Y %m %d %H %M %S").date():
            day1.append(city['text'].values[i])
        if date_compare[1]==datetime.datetime.strptime(dates[i],"%Y %m %d %H %M %S").date():
            day2.append(city['text'].values[i])
        if date_compare[2]==datetime.datetime.strptime(dates[i],"%Y %m %d %H %M %S").date():
            day3.append(city['text'].values[i])

    day1_Sent = process(day1)
    day2_Sent = process(day2)
    day3_Sent = process(day3)
    day1sent = overallt(day1_Sent)
    day2sent = overallt(day2_Sent)
    day3sent = overallt(day3_Sent)
    print("Day1"+ str(day1sent))
    print("Day2"+ str(day2sent))
    print("Day3"+ str(day3sent))
```

```python
def PreProcessing():
    df=pd.read_csv("https://raw.githubusercontent.com/tubbytani/IBM-hack-challenge/master/EXTRACTED.csv?token=ALEGRVODNLMSUKABCQIRFTS67WZ5Q")
    #df=pd.read_csv("EXTRACTED.CSV",encoding = "utf-8")
    df = df.applymap(lambda s:s.lower() if type(s) == str else s)
    df = df.applymap(lambda s:clean(s) if type(s) == str else s)
    df = df.applymap(lambda s:remove_stopwords(s) if type(s) == str else s)
    df = df.apply(lambda s:Textblob(s).correct() if type(s) == str else s)
    new= df[df["location"].str.contains("delhi")]
    new2=df[df["location"].str.contains("bengal")]
    new3= df[df["location"].str.contains("tamil")]
    new4= df[df["location"].str.contains("mumbai")]
    new.to_csv("delhi.csv")
    new2.to_csv("bengal.csv")
    new3.to_csv("tamil.csv")
    new4.to_csv("mumbai.csv")
    df.to_csv("overall.csv")
```

```python
def process(tweets):
    x =[]
    for i in range(len(tweets)):
        x.append(get_tweet_sentiment(tweets[i]))
    return x

def overallt(tweetsent):
    overallpos=0
    overallneg=0
    overallneutral=0
    for i in tweetsent:
        if i > 0:
            overallpos+=1
        if i < 0:
            overallneg+=1
        if i==0:
            overallneutral+=1
    listcount =[overallpos,overallneg,overallneutral]
    return listcount
    
```
